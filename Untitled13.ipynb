```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
df= pd.read_csv("supermarket_sales - Sheet1.csv")
df.head()
```

```python
df = pd.read_csv("supermarket_sales - Sheet1.csv", parse_dates=['Date'])
```

```python
df.head(2)
```

```python
# 1. Basic cleaning
df.columns = [c.strip().replace(' ', '_').replace('%','pct').lower() for c in df.columns]
```

```python
df.head()
```

```python
# convert types
df['time'] = pd.to_datetime(df['time'], format='%H:%M').dt.time
```

```python
# derived columns
df['sales'] = df['unit_price'] * df['quantity']
```

```python
# confirm gross income present
if 'gross_income' not in df.columns:
    df['gross_income'] = df['sales'] - df.get('cogs', 0)
```

```python
df.head(2)
```

```python
profit_by_product = df.groupby('product_line').agg(
    total_sales=('sales', 'sum'),
    total_gross_income=('gross_income', 'sum'),
    avg_gross_margin_pct=('gross_margin_percentage', 'mean'),
    units_sold=('quantity', 'sum'),
    transactions=('invoice_id', 'nunique')
).sort_values('total_gross_income', ascending=False)
profit_by_branch = df.groupby('branch').agg(
    total_sales=('sales', 'sum'),
    total_gross_income=('gross_income', 'sum'),
    units_sold=('quantity', 'sum')
).sort_values('total_sales', ascending=False)
```

```python
print(profit_by_product)
```

```python
print(profit_by_branch)
```

```python
# 3. Seasonality - monthly sales
df['month'] = df['date'].dt.to_period('M')
monthly = df.groupby('month').agg(total_sales=('sales','sum'), total_transactions=('invoice_id','nunique')).reset_index()
monthly['month'] = monthly['month'].dt.to_timestamp()
```

```python
# 4. Top and bottom performing product lines
top_products = profit_by_product.head(5)
bottom_products = profit_by_product.tail(5)
```

```python
# 5. Correlation analysis (example: units sold vs gross income)
corr = df[['quantity','gross_income','unit_price']].corr()
corr
```
